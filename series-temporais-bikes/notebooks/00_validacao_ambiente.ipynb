```python
df_teste = df.copy()

df_teste["datetime"] = pd.to_datetime(df_teste["datetime"])

print("Primeira data:", df_teste["datetime"].min())
print("Última data:", df_teste["datetime"].max())
```

```python
fig = grafico_linha_padrao(
    df=df_teste,
    x="datetime",
    y="count",
    titulo="Demanda de bicicletas ao longo do tempo",
    labels={
        "datetime": "Data e hora",
        "count": "Demanda de bicicletas"
    },
    altura=500
)

fig.show()
```
